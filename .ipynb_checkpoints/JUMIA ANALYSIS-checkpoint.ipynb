# First Part:
+ "Jumia launched in 2012 and its largest category in terms of NMV is phones & consumer electronics. Part of the challenge of becoming a generalist online store was to lower the average order value in 2018. What levers would you pull to lower the average order value? How do you sustain this from a profitability standpoint?"
+ Solution: The average order value depends on the turnover and the number of orders; to reduce it (Mathematical point of view) is to increase the number of orders, since the largest category in terms of NMV is that of telephones and electronics and which require a high turnover, therefore from a point of view of profitability increase turnover (with a lower rate; for example less expensive categories) and increase more the number of orders: Display product recommendations; Offer free shipping; leads to reduce the AOV.

# Q7 : What products are more likely to be cancelled? What would you recommend to lower the cancellation rate?Â 

```python
q71 = pd.DataFrame(df.groupby(['Product_category'],as_index=False)['canceled_transactions'].sum())

q72 = pd.DataFrame(df.groupby(['Product_category'],as_index=False)['gross_transactions'].sum()) 
```

```python
q71.drop('Product_category',axis=1,inplace=True)
frames = [q72, q71]

result = pd.concat(frames,axis=1,ignore_index=False)
```

```python
result.head()
```

```python
#Rounding the value of rate
result['rate'] = round((result['canceled_transactions']/result['gross_transactions'])*100,2)
```

```python
#Sorting the dataFrame 
result.sort_values(by='rate',ascending=False,inplace=True)
```

```python
result.head(10)
```

```python
%matplotlib inline 
plt.figure(figsize=(5,30))
sns.barplot(x='rate',y='Product_category',data=result)
plt.title(' Cancellation-rate / Product_category')
```

<h3 style="color: blue;">  Answer 7 : As the Plot is showing : there are many different products that was totaly cancelled such as : Women's Eastern Wear/Water Sports/Prepared Foods/Real Estate... . In my oppinion i would recommend to totaly delete some categories for example the "Religious & Ceremonial" and  "Water Sports" had only one Transactions !! there will not be a good benifit and for our case we consider it like outlieres </h3>

# Q9. Which category has the best net/gross ratio. What do you think this means?

```python
q91 = pd.DataFrame(df.groupby(['Product_category'],as_index=False)['net_transactions'].sum())
q92 = pd.DataFrame(df.groupby(['Product_category'],as_index=False)['gross_transactions'].sum()) 
q91.drop('Product_category',axis=1,inplace=True)
frames = [q92, q91]
q9 = pd.concat(frames,axis=1,ignore_index=False)
```

```python
q9['ratio'] = round((q9['net_transactions']/q9['gross_transactions'])*100,2)
```

```python
best_ratio = q9.sort_values(by='ratio',ascending=False)
```

```python
best_ratio = best_ratio[(best_ratio['ratio'] > 0) & (best_ratio['net_transactions'] > 100)]
```

```python
best_ratio.sort_values(by='gross_transactions',ascending=False)
```

<h3 style="color: blue;"> Answer Q9: The Answer will not be a number, the same as the questions before, the ratio callculated on all the Product-Categories will not give as a good result because  the gross_transactions are too different; If we focuce on the top 5 CATS-gross_transactions the best one is "Women's Clothing" </h3>
